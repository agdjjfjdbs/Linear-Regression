```python
data_set = pd.read_csv('Admission_Predicttt.csv')
x=data_set.iloc[:,6:7].values
x
```

```python
y=data_set.iloc[:,8:9].values
```

take care missing data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan , strategy='mean')
imputer.fit(x)
x = imputer.transform(x)
```

```python
print(x)
```

Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data_set["CGPA_encoded"] = encoder.fit_transform(data_set["CGPA"])
data_set["Admission_encoded"] = encoder.fit_transform(data_set["Admission"])
print(data_set["CGPA_encoded"])
print(data_set["Admission_encoded"])
```

visulizing data

```python
fig = plt.figure(figsize=(12, 6))
CGPA = fig.add_subplot(121)
admission  = fig.add_subplot(122)

CGPA.hist(data_set.CGPA, bins=10)
CGPA.set_xlabel('CGPA')
CGPA.set_title("Histogram of CGPA")

admission.hist(data_set.Admission , bins=10)
admission.set_xlabel('Admission')
admission.set_title("Histogram of Admission ")

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x)
x = sc.transform(x)
x
```

Spiliting the data set into training data and test data

```python
from sklearn.model_selection import train_test_split
```

```python
x.shape, y.shape
```

```python

x_train, x_test,y_train, y_test=train_test_split(x,y,test_size=0.20, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
modle = LinearRegression()
```

visualizing the training test result

```python
plt.scatter(x_train,y_train, color='red')
plt.plot(x_train, modle.predict(x_train), color='blue')
plt.title('graduate admission for train set')
plt.xlabel('gpa')
plt.ylabel('graduate admission')
```

visulizing the test data results

```python
plt.scatter(x_test,y_test, color='red')
plt.plot(x_train, modle.predict(x_train), color='blue')
plt.title('graduate admission for test set')
plt.xlabel('gpa')
plt.ylabel('graduate admission')
plt.show()
```

```python
y_train_pred = modle.predict(x_train)
print(y_train_pred)
```

 R square and MAE / RMSE

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
plt.scatter(x_train, y_train)
plt.scatter(x_train, y_train_pred)
```

```python
print('R2 = ', r2_score(y_train, y_train_pred))
print('MAE = ', mean_absolute_error(y_train, y_train_pred))
print('MRSE = ', mean_squared_error(y_train, y_train_pred))
```

predicting the test set results

```python
plt.scatter(x_test, y_test)
plt.scatter(x_test, y_test_pred)
```

```python
print('R2 = ', r2_score(y_test, y_test_pred))
print('MAE = ', mean_absolute_error(y_test, y_test_pred))
print('MRSE = ', mean_squared_error(y_test, y_test_pred))
```
